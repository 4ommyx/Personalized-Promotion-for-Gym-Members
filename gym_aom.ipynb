```python
## Select Feature (Feature Importance)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

## Pred (Experience_Level)
X = df.drop(columns=['Experience_Level'])
y = df['Experience_Level']

# trans categorical features
le = LabelEncoder()
X['Gender'] = le.fit_transform(X['Gender'])
X['Workout_Type'] = le.fit_transform(X['Workout_Type'])

X_raw = X.copy()

### dataset to feature importance ###
X_train_df, X_test_df, y_train_df, y_test_df = train_test_split(X_raw, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_df, y_train_df)

# Feature importance
importances = rf_model.feature_importances_
feature_names = X_raw.columns

df_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(12, 10))
plt.barh(df_importance['Feature'], df_importance['Importance'])
plt.gca().invert_yaxis()
plt.title('Feature Importance for Predicting Experience_Level')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
import time
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import ParameterGrid

# Select top 4 Feature Selection
top_features = df_importance['Feature'].tolist()[:4]
X_train_top = X_train_df[top_features]
X_test_top = X_test_df[top_features]

# Scaling
scaler = StandardScaler()
X_train_top_scaled = scaler.fit_transform(X_train_top)
X_test_top_scaled = scaler.transform(X_test_top)

# train model (KNN, D-tree, SVM, RF)
results = []

# KNN
knn = KNeighborsClassifier(weights='distance')
param_grid_knn = {'n_neighbors': range(1, 16)}
grid_knn = GridSearchCV(knn, param_grid_knn, cv=5, scoring='accuracy')
grid_knn.fit(X_train_top_scaled, y_train_df)
best_knn = grid_knn.best_estimator_

start_train = time.perf_counter()
best_knn.fit(X_train_top_scaled, y_train_df)
end_train = time.perf_counter()

start_pred = time.perf_counter()
y_pred_knn = best_knn.predict(X_test_top_scaled)
end_pred = time.perf_counter()

results.append({
    'Model': 'KNN',
    'Accuracy': accuracy_score(y_test_df, y_pred_knn),
    'Train Time': end_train - start_train,
    'Predict Time': end_pred - start_pred,
    'Report': classification_report(y_test_df, y_pred_knn, output_dict=True),
    'y_pred': y_pred_knn
})

# Random Forest
rf = RandomForestClassifier(random_state=42)

start_train = time.perf_counter()
rf.fit(X_train_top, y_train_df)
end_train = time.perf_counter()

start_pred = time.perf_counter()
y_pred_rf = rf.predict(X_test_top)
end_pred = time.perf_counter()

results.append({
    'Model': 'Random Forest',
    'Accuracy': accuracy_score(y_test_df, y_pred_rf),
    'Train Time': end_train - start_train,
    'Predict Time': end_pred - start_pred,
    'Report': classification_report(y_test_df, y_pred_rf, output_dict=True),
    'y_pred': y_pred_rf
})

# Decision Tree
dt = DecisionTreeClassifier(random_state=42)

start_train = time.perf_counter()
dt.fit(X_train_top, y_train_df)
end_train = time.perf_counter()

start_pred = time.perf_counter()
y_pred_dt = dt.predict(X_test_top)
end_pred = time.perf_counter()

results.append({
    'Model': 'Decision Tree',
    'Accuracy': accuracy_score(y_test_df, y_pred_dt),
    'Train Time': end_train - start_train,
    'Predict Time': end_pred - start_pred,
    'Report': classification_report(y_test_df, y_pred_dt, output_dict=True),
    'y_pred': y_pred_dt
})

# SVM
svm = SVC(kernel='linear')

start_train = time.perf_counter()
svm.fit(X_train_top_scaled, y_train_df)
end_train = time.perf_counter()

start_pred = time.perf_counter()
y_pred_svm = svm.predict(X_test_top_scaled)
end_pred = time.perf_counter()

results.append({
    'Model': 'SVM',
    'Accuracy': accuracy_score(y_test_df, y_pred_svm),
    'Train Time': end_train - start_train,
    'Predict Time': end_pred - start_pred,
    'Report': classification_report(y_test_df, y_pred_svm, output_dict=True),
    'y_pred': y_pred_svm
})

# Summary all model
results_df = pd.DataFrame({
    'Model': [r['Model'] for r in results],
    'Accuracy': [r['Accuracy'] for r in results],
    'Train Time (s)': [r['Train Time'] for r in results],
    'Predict Time (s)': [r['Predict Time'] for r in results]
})

print("\n=== Model Performance Summary ===")
print(results_df)

for r in results:
    print(f"\n--- {r['Model']} Classification Report ---")
    print(classification_report(y_test_df, r['y_pred']))
```

```python
# Select the top x_features based on feature importance
features_sorted = df_importance['Feature'].tolist()

### training loop (for hyperparameter-tuning) - > (n_feat, n_estimators, max_depth)
log_results = []

param_grid_rf = {
    'n_estimators': [10, 50, 100],
    'max_depth': [5, 10, 20]
}

param_combinations = list(ParameterGrid(param_grid_rf))

for i in range(1, len(features_sorted) + 1):
    selected_features = features_sorted[:i]

    # select x train
    X_sel_train = X_train_df[selected_features]
    X_sel_test = X_test_df[selected_features]

    # scaler by StandardScaler
    scaler = StandardScaler()
    X_sel_train_scaled = scaler.fit_transform(X_sel_train)
    X_sel_test_scaled = scaler.transform(X_sel_test)

    for params in param_combinations:
            rf_model = RandomForestClassifier(**params, random_state=42)
            rf_model.fit(X_sel_train_scaled, y_train_df)

            y_pred = rf_model.predict(X_sel_test_scaled)
            acc = accuracy_score(y_test_df, y_pred)

            log_results.append({
                'Top_Features_Used': i,
                'Selected_Features': selected_features,
                'Params': params,
                'Accuracy': acc
            })

            print(f"Features : Top {i} | Params: {params} | Accuracy: {acc:.4f}")
    print('---'*30)
```

```python
# trans log_results to DataFrame
df_log = pd.DataFrame(log_results)

# Extract hyperparameters
df_log['n_estimators'] = df_log['Params'].apply(lambda x: x['n_estimators'])
df_log['max_depth'] = df_log['Params'].apply(lambda x: x['max_depth'])

# Prepare 3D surface plot
x = df_log['Top_Features_Used']
y = df_log['n_estimators']
z = df_log['max_depth']
acc = df_log['Accuracy']

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(x, y, z, c=acc, cmap='viridis', s=50)
ax.set_xlabel('Top Features Used')
ax.set_ylabel('n_estimators')
ax.set_zlabel('max_depth')
ax.set_title('Accuracy by Feature Count and Hyperparameters')

cbar = plt.colorbar(sc, ax=ax, shrink=0.5, aspect=10)
cbar.set_label('Accuracy')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

### train model (grid scarch + top 4 feat)
def rf_select_feat(feat = 3):

    # rand param
    n_estimators_list = [10, 50, 100, 150, 200]
    max_depth_list = [5, 10, 15, 20, 25]

    # select top features
    features_sorted = df_importance.sort_values(by='Importance', ascending=False)['Feature'].tolist()
    selected_features = features_sorted[:feat]

    X_sel_train = X_train_df[selected_features]
    X_sel_test = X_test_df[selected_features]

    scaler = StandardScaler()
    X_sel_train_scaled = scaler.fit_transform(X_sel_train)
    X_sel_test_scaled = scaler.transform(X_sel_test)

    results = []

    for max_depth in max_depth_list:
        for n_estimators in n_estimators_list:
            model = RandomForestClassifier(n_estimators=n_estimators,
                                          max_depth=max_depth,
                                          random_state=42)
            model.fit(X_sel_train_scaled, y_train_df)
            y_pred = model.predict(X_sel_test_scaled)
            acc = accuracy_score(y_test_df, y_pred)

            results.append({
                'n_estimators': n_estimators,
                'max_depth': max_depth,
                'Accuracy': acc,
                'Model': model
            })
            # print(f"n_estimators={n_estimators}, max_depth={max_depth} => Accuracy: {acc:.4f}")

    df_results = pd.DataFrame(results)

    # find model top acc สูงสุด
    top_acc_row = df_results.loc[df_results['Accuracy'].idxmax()]

    # Plot Line Chart
    plt.figure(figsize=(10, 6))
    for depth in max_depth_list:
        subset = df_results[df_results['max_depth'] == depth]
        plt.plot(subset['n_estimators'], subset['Accuracy'], marker='o', label=f'max_depth={depth}')

    plt.title(f"Accuracy vs. n_estimators (Top {feat} Features)")
    plt.xlabel("n_estimators")
    plt.ylabel("Accuracy")
    plt.ylim(0, 1)
    plt.grid(True)
    plt.legend(title="max_depth")
    plt.ylim(0.8, 1)  # startwith 0.8
    plt.tight_layout()
    plt.show()

    print('---'*30)
    print('Best Model')
    best_model = top_acc_row['Model']
    print(f"Top Accuracy: {top_acc_row['Accuracy']:.4f}")
    print(f"Best parameters -> n_estimators: {top_acc_row['n_estimators']}, max_depth: {top_acc_row['max_depth']}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test_df, best_model.predict(X_sel_test_scaled)))
    print('---'*30)

    # Confusion Matrix
    cm = confusion_matrix(y_test_df, best_model.predict(X_sel_test_scaled))
    print(cm)

    # heatmap (cm)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix (Heatmap)')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.tight_layout()
    plt.show()

    return df_results

# call func
df_log_grid = rf_select_feat(3)
# # Confusion Matrix
#     print("\nConfusion Matrix:")
#     cm = confusion_matrix(y_test_df, best_model.predict(X_sel_test_scaled))
#     print(cm)
```

```python
## Feature Engineering

# Heart Rate Ratio (Avg_BPM / Max_BPM) -> ความหนักของการออกกำลังกาย
df['Heart_Rate_Ratio'] = df['Avg_BPM'] / (df['Max_BPM'] + 1e-6)

# Weight to Height Ratio
df['Weight_to_Height_Ratio'] = df['Weight (kg)'] / (df['Height (m)'] + 1e-6)

# Fat Percentage to Water Intake Ratio
df['Fat_Percentage_to_Water_Ratio'] = df['Fat_Percentage'] / (df['Water_Intake (liters)'] + 1e-6)

# Consistency_Score -> (ความสม่ำเสมอในการออกกำลัง)
df['Consistency_Score (hours/week)'] = df['Workout_Frequency (days/week)'] * df['Session_Duration (hours)']

# Intensity_Score -> เทียบจาก Calories_Burned
df['Intensity_Score (cal/hours)'] = df['Calories_Burned'] / (df['Session_Duration (hours)'] + 1e-6)

df.head()
```

```python
from sklearn.metrics import silhouette_score
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
from sklearn.cluster import KMeans


selected_features = ['Intensity_Score (cal/hours)', 'Fat_Percentage_to_Water_Ratio', 'Consistency_Score (hours/week)']
X = df[selected_features]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Use matric (find best k)
inertia = []
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)
    cluster_labels = kmeans.fit_predict(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, cluster_labels))

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia, marker='o')
plt.title('KMeans Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Scores for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.tight_layout()
plt.show()
print(inertia)
print(silhouette_scores)
```
